```python
list_spec = [spec_0p25ng, spec_1ng, spec_5ng, spec_20ng, spec_50ng, spec_200ng, spec_500ng, spec_1000ng]
list_insol = [insol_0p25ng, insol_1ng, insol_5ng, insol_20ng, insol_50ng, insol_200ng, insol_500ng, insol_1000ng]
list_pac = [pac_0p25ng, pac_1ng, pac_5ng, pac_20ng, pac_50ng, pac_200ng, pac_500ng, pac_1000ng]
```

```python
tmp_df_list_pg_spec = []
tmp_df_list_prec_spec = []
for el in list_spec:
    tmp_df = el[el['PG.Q.Value']< 0.01]
    tmp_df1 = pd.pivot_table(tmp_df, columns = 'Run', index = 'Protein.Group', values= 'PG.MaxLFQ')
    tmp_df_list_pg_spec.append(tmp_df1)
    tmp_df2 = pd.pivot_table(tmp_df, columns = 'Run', index = 'Precursor.Id', values= 'Precursor.Normalised')
    tmp_df_list_prec_spec.append(tmp_df2)


#####
tmp_df_list_pg_insol = []
tmp_df_list_prec_insol = []
for el in list_insol:
    tmp_df = el[el['PG.Q.Value']< 0.01]
    tmp_df1 = pd.pivot_table(tmp_df, columns = 'Run', index = 'Protein.Group', values= 'PG.MaxLFQ')
    tmp_df_list_pg_insol.append(tmp_df1)
    tmp_df2 = pd.pivot_table(tmp_df, columns = 'Run', index = 'Precursor.Id', values= 'Precursor.Normalised')
    tmp_df_list_prec_insol.append(tmp_df2)

#####
tmp_df_list_pg_pac = []
tmp_df_list_prec_pac = []
for el in list_pac:
    tmp_df = el[el['PG.Q.Value']< 0.01]
    tmp_df1 = pd.pivot_table(tmp_df, columns = 'Run', index = 'Protein.Group', values= 'PG.MaxLFQ')
    tmp_df_list_pg_pac.append(tmp_df1)
    tmp_df2 = pd.pivot_table(tmp_df, columns = 'Run', index = 'Precursor.Id', values= 'Precursor.Normalised')
    tmp_df_list_prec_pac.append(tmp_df2)
```

```python
cv_list_pg_spec = []
for df in tmp_df_list_pg_spec:
    linear_values = df.iloc[:,:4]

    n_valid = linear_values.notna().sum(axis=1)
    
    cv = linear_values.std(axis=1) / linear_values.mean(axis=1)
    cv = cv.replace(0, np.nan)

    cv[n_valid < 2] = np.nan
    
    cv_list_pg_spec.append(cv)

cv_list_prec_spec = []
for df in tmp_df_list_prec_spec:
    linear_values = df.iloc[:,:4]

    n_valid = linear_values.notna().sum(axis=1)
    
    cv = linear_values.std(axis=1) / linear_values.mean(axis=1)
    cv = cv.replace(0, np.nan)

    cv[n_valid < 2] = np.nan
    
    cv_list_prec_spec.append(cv)
```

```python
cv_list_pg_insol = []
for df in tmp_df_list_pg_insol:
    linear_values = df.iloc[:,:4]

    n_valid = linear_values.notna().sum(axis=1)
    
    cv = linear_values.std(axis=1) / linear_values.mean(axis=1)
    cv = cv.replace(0, np.nan)

    cv[n_valid < 2] = np.nan
    
    cv_list_pg_insol.append(cv)

cv_list_prec_insol = []
for df in tmp_df_list_prec_spec:
    linear_values = df.iloc[:,:4]

    n_valid = linear_values.notna().sum(axis=1)
    
    cv = linear_values.std(axis=1) / linear_values.mean(axis=1)
    cv = cv.replace(0, np.nan)

    cv[n_valid < 2] = np.nan
    
    cv_list_prec_insol.append(cv)
```
